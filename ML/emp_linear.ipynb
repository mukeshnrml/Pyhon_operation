```python
#x = df.drop(['First Name','Gender','Start Date','Bonus %','Senior Management','Team'])
x = df[['Salary']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Build Model
reg = LinearRegression()
```

```python
# Training model
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```
