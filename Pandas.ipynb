```python
import pandas as pd 
dff = pd.read_csv('mark.csv')
dff
```

```python
dff['Total'] = dff['Hindi']+dff['English']+dff['Science']
dff
```

```python
dff.head()
```

```python
dff.loc[(dff.Hindi == 80)]
```

```python
dff.loc[(dff.English >= 90) | (dff.Science >=80)]
```

```python
dff['Percentage']= (dff['Total']/500)*100
dff
```

```python
dff['Grade']= 'Pass'
dff
```

```python
dff.loc[dff['Percentage'] >=45, ['Grade']] = 'First'
dff.loc[dff['Percentage'] <=45, ['Grade']] = 'Second'
dff
```
